```python
import xml.etree.ElementTree as ET
SIG_STR = "sig"
FIELD_STR = "field"
def parse_instance():
    tree = ET.parse("test.xml")
    root = tree.getroot()
    instance,source = list(root)
    forge_file_source = source.attrib["content"]
    sig_field_and_skolems = list(instance)
    sigs,fields = [],[]
    for sig_or_field in sig_field_and_skolems:
        tag = sig_or_field.tag
        if tag == SIG_STR:
            sigs.append(sig_or_field)
        elif tag == FIELD_STR:
            fields.append(sig_or_field)
    return sigs,fields
parse_instance()
```
